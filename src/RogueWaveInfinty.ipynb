First step: load the distributed packages and then their library

```julia
using Distributed
addprocs(8)
@everywhere using RogueWaveInfiniteNLS
```

```julia
# Computations
@everywhere Xlist = collect(-20:0.1:20)
@everywhere atest = 1
@everywhere btest = -1
@everywhere compute_psi_slice = X->psi(X,0,atest,btest,1)
outpsiX = pmap(compute_psi_slice, Xlist)
```

```julia
#   Convert to absolute values
# Plots and CSV output
num_points = length(Xlist)
abs_outpsiX = zeros(length(Xlist))
for j in 1:num_points
    abs_outpsiX[j] = abs(outpsiX[j])
end
using Plots, CSV, DataFrames
df = DataFrame(X = Xlist, PsiAbs = abs_outpsiX)
CSV.write("RogueWaveInfiniteNLS_Output.csv", df)
plot(Xlist, abs_outpsiX, xlabel="X", ylabel="|(X,0)|", title=" at t=0", legend=true)  
```

```julia
using RogueWaveInfiniteNLS
Xlist = collect(-16:1:16)
outpsiX = zeros(length(Xlist))
for j in 1:length(Xlist)
    outpsiX[j] =  abs(psi(Xlist[j], 0, 1, 1, 1)) #(x,t,a,b,beta)
end
println(outpsiX)
```
